Now for a competitive pricing of the Dishes, we analyse and compare the prices at various other restaurants. We will compare our prices with the AVERAGE MARKET PRICE for that particular dish. We also compare it with the MEDIAN PRICE and the MODE PRICE. It should be always be close to these central identities specially the MEAN and MEDIAN. Also, if there isn't a unique most frequently occuring value, we take the lowest MODE into consideration.

NOTE: If our price for a certain dish is higher than all the three MEAN, MEDIAN AND MODE and that particular dish is not the most sold or loved dish of our Restaurant, then we may need to consider lowering the prices. The same goes if our dish is at too low cost.

The Prices are displayed in the order.
GOLI'S PRICE, MEAN, MEDIAN, MODE

```python
import statistics
d_pricing = {}
for a in Goli_dishes:
    Goli_ind = []
    val  = 0
    ind = []
    l = []
    Goli_ind = Golidf.index[Golidf['Menu Item'] == a].tolist()
    Goli_Price = Golidf.get_value(Goli_ind[0],'Rate')
    ind  = df1.index[df1['Menu Item'] == a].tolist()
    for t in ind:
        if((df1.get_value(t,'Rate') == '')) :
           
            ind.remove(t)
        else:
            
            try:
                if(int(df1.get_value(t,'Rate'))<1):
                    ind.remove(t)
                
                else:
                    try:
                        
                        l.append(int(ef.get_value(t,'Rate')))
                
                        val = val + int(ef.get_value(t,'Rate'))
                    except ValueError:
                        ind.remove(t)
            except ValueError:
                ind.remove(t)
    median = statistics.median(l)
    mean = statistics.mean(l)
    try:
        mode = statistics.mode(l)
    except:
        mode = min(l)
    d_pricing[a] = [Goli_Price,mean,median,mode]
print('Goli,s Price,Mean,Median,Mode')
d_pricing
        
```
